## Data cleaning and prepocessing

```python
# Dropping irrelevant columns: 
df.drop(columns=['adult', 'imdb_id', 'original_title', 'video', 'homepage'], inplace =True)
```

```python
df.head()
```

```python
# Extracting data from the JSON-like columns
import ast

def clean_json_columns(df):
    """
    Function to extract data from JSON-like columns
    """
    # Create a copy so we don't modify the original
    processed_df = df.copy()
    
    # Handle genres column
    if 'genres' in df.columns:
        def extract_genres(value):
            if pd.isna(value) or value == '':
                return ''
            try:
                # Convert from string if needed
                if isinstance(value, str):
                    value = ast.literal_eval(value)
                # Extract names and join with pipe
                return '|'.join([item['name'] for item in value if 'name' in item])
            except:
                return ''
        
        processed_df['genres'] = processed_df['genres'].apply(extract_genres)
    
    # Handle collection
    if 'belongs_to_collection' in df.columns:
        def extract_collection(value):
            if pd.isna(value) or value == '' or value is None:
                return ''
            try:
                if isinstance(value, str):
                    value = ast.literal_eval(value)
                return value['name'] if value and 'name' in value else ''
            except:
                return ''
                
        processed_df['belongs_to_collection'] = processed_df['belongs_to_collection'].apply(extract_collection)
    
    # Handle all other list columns the same way
    columns_to_process = ['production_countries','production_companies','spoken_languages']

    
    for col in columns_to_process:
        if col in df.columns:
            processed_df[col] = processed_df[col].apply(extract_genres)
    
    return processed_df


# Clean the JSON-like columns
cleaned_df = clean_json_columns(df)

```

```python
# Inspecting extracted columns using value_counts() to identify anomalies.
for extract in ['belongs_to_collection', 'genres', 'production_countries','production_companies', 'spoken_languages']:
    print(f"{cleaned_df[extract].value_counts()}\n")
```

## Handling missing and Incorrect data

```python
cleaned_df[['budget', 'id', 'popularity']].dtypes
```

```python
# Change 'release_date' to datetime
cleaned_df['release_date']=pd.to_datetime(cleaned_df['release_date'])
cleaned_df.info()
```

## Replacing unrealistic data

```python
# Budget/Revenue/Runtime = 0 → Replacing with NaN 
cleaned_df[['budget','revenue','runtime']]=cleaned_df[['budget','revenue','runtime']].apply(lambda x: x.replace(0,np.nan))
cleaned_df[['budget','revenue','runtime']]
```

```python
# Convert 'budget' and 'revenue' to million USD
cleaned_df[['budget (million usd)','revenue (million usd)']]=\
cleaned_df[['budget','revenue']].apply(lambda x: round(x/10000000,2))

cleaned_df.columns
```

```python
# drop revenue and budget
cleaned_df.drop(columns=['revenue','budget'], inplace = True)
```

```python
cleaned_df.columns
```

### Working on the dataset containing 'cast', 'cast_size', 'director', 'crew_size' columns and then merge with the movie dataset

```python
# Importing the data containing 'cast', 'cast_size', 'director', 'crew_size' columns
credits_df =pd.read_csv("../data/extracted_df1.csv")
credits_df
```

```python
# dropping the row with with zero value
credits_df.drop(0, inplace=True)
```

```python
# stripping the '[]' off the director column
credits_df['director']=credits_df['director'].str.strip("[],'' ")
credits_df
```

```python
# merging the two dataframes

df_merge=pd.merge(cleaned_df,credits_df, left_on='id',right_on='movie_id')
df_merge.head()
```
